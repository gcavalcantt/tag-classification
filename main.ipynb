```
questions.Tags.unique()
```

```
tag_list = list()
for tags in questions.Tags.unique():
    for tag in tags.split():
        if tag not in tag_list:
            tag_list.append(tag)

print(tag_list)
```

```
node_js = list()
for tag_line in questions.Tags:
    if "node.js" in tag_line:
        node_js.append(1)
    else:
        node_js.append(0)
questions["node.js"] = node_js

questions
```

```
def new_column(tag_list, dataframe, tag_column_name):
    for tag in tag_list:
        column = list()
        for tag_line in dataframe[tag_column_name]:
            if tag in tag_line:
                column.append(1)
            else:
                column.append(0)
        dataframe[tag] = column

new_column(tag_list, questions, "Tags")
questions.sample(10)
```

# Vectorization

Applying TfidfVectorizer.

```
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000, max_df=0.85)
vectorizer
```

```
vectorizer.fit(questions.Perguntas) # Questions
train_questions_tfidf = vectorizer.transform(train_questions)
test_questions_tfidf = vectorizer.transform(test_questions)

print(train_questions_tfidf.shape)
print(test_questions_tfidf.shape)
```

# One-vs-Rest (OvR)
Also known as One-vs-All, it is a classification strategy where the multiclass problem is divided into multiple binary problems. For each class, a binary classifier is trained to distinguish whether a sample belongs to that class (with "rest" referring to all the other classes).

```
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MultiLabelBinarizer
import numpy as np
```

```
# Binarizing the tags
mlb = MultiLabelBinarizer()
train_tags_binarized = mlb.fit_transform(train_tags)
test_tags_binarized = mlb.transform(test_tags)
```

```
# Training the classifier
logistic_regression = LogisticRegression()
onevsrest_classifier = OneVsRestClassifier(logistic_regression)
onevsrest_classifier.fit(train_questions_tfidf, train_tags_binarized)
```

# Hamming Loss
A metric used to evaluate the performance of multilabel classification models. It measures the fraction of incorrectly predicted labels relative to the total number of labels.

```
from sklearn.metrics import hamming_loss

onevsrest_prediction = onevsrest_classifier.predict(test_questions_tfidf)
hamming_loss_onevsrest = hamming_loss(test_tags_array, onevsrest_prediction)

print("Hamming Loss: {0:.2f}".format(hamming_loss_onevsrest))
```

Performing the final trainings and checking predictive results.

```
from skmultilearn.problem_transform import ClassifierChain
from skmultilearn.problem_transform import BinaryRelevance
from skmultilearn.adapt import MLkNN
```

```
# Initializing the ClassifierChain with logistic regression
chain_classifier = ClassifierChain(logistic_regression)

# Training the classifier
chain_classifier.fit(train_questions_tfidf, train_tags_array)

# Evaluating the model
chain_result = chain_classifier.score(test_questions_tfidf, test_tags_array)
chain_prediction = chain_classifier.predict(test_questions_tfidf)
chain_hamming_loss = hamming_loss(test_tags_array, chain_prediction)

print("Hamming Loss: {0:.2f}".format(chain_hamming_loss))
print("Result: {0:.2f}%".format(chain_result * 100))
```

```
# Initializing the BinaryRelevance with logistic regression
br_classifier = BinaryRelevance(logistic_regression)

# Training the binariser
br_classifier.fit(train_questions_tfidf, train_tags_array)

# Evaluating the model
br_result = br_classifier.score(test_questions_tfidf, test_tags_array)
br_prediction = br_classifier.predict(test_questions_tfidf)
br_hamming_loss = hamming_loss(test_tags_array, br_prediction)

print("Hamming Loss: {0:.2f}".format(br_hamming_loss))
print("Result: {0:.2f}%".format(br_result * 100))
```

```
classification_results = pd.DataFrame()

# Creating columns for the result models
classification_results["questions"] = test_questions.values
classification_results["actual tags"] = list(test_tags)
classification_results["BR"] = list(br_prediction.toarray())
classification_results["chain"] = list(chain_prediction.toarray())

classification_results
```
